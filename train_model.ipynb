```
import pandas as pd
from sklearn.linear_model import LogisticRegression
import sys
sys.path.append('/content/drive/My Drive/Colab Notebooks/ML_Model_Api')
# create df
path = '/content/drive/My Drive/Colab Notebooks/ML_Model_Api'
#Read csv using pandas
train = pd.read_csv(path+'/titanic.csv')
# train = pd.read_csv('titanic.csv') # change file path
# drop null values
train.dropna(inplace=True)
# features and target
target = 'Survived'
features = ['Pclass', 'Age', 'SibSp', 'Fare']
# X matrix, y vector
X = train[features]
y = train[target]
# model 
model = LogisticRegression()
model.fit(X, y)
model.score(X, y)
```

```
# dump the model 
import pickle
pickle.dump(model, open(path+'/model.pkl', 'wb'))
```
